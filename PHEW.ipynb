```python
import os

%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import astropy.io.ascii as at
import astropy.io.fits as fits

import PHEW as phew
```

```python
mdm = at.read("new_spectra.lst")
mdm = mdm["filename"]
```

```python
with fits.open(mdm[0]) as hdu:
    print(hdu.info())
    print(hdu[0].data)
    for i in range(4):
        plt.plot(hdu[0].data[i][0],alpha=0.5)
plt.yscale("log")
```

```python
with fits.open(mdm[0]) as hdu:
    print(hdu.info())
#     print(hdu[0].header)
    flux = hdu[0].data[0][0]
    hdr = hdu[0].header
    hdr["WAT1_001"] = hdr["WAT1_001"].replace("angstroms", "Angstrom")
    
    
#     hdr.
    dw = hdr["CD1_1"]
    l = len(flux)
    minpix = hdr["CRPIX1"]
    minw = hdr["CRVAL1"]
    maxw = hdr["CRVAL1"] + (l-minpix)*dw
    print(minpix)
    startpix = int(-1*minpix)
    print(startpix)
    w = np.arange(minw,maxw,dw)[startpix:]
    print(len(flux),len(w))
    
    fits.writeto("test.fits",data=hdu[0].data,header=hdr,overwrite=True)
    
plt.plot(w, flux)
# plt.xlim(6400,6700)
```
